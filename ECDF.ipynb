```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
e_d = pd.read_csv('ECDFdata.csv')
```

```python
def ECDF(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1 , n+1)/n
    return x , y
x , y = ECDF(e_d.inch)
```

```python
plt.scatter(x , y)
plt.show()
```
